## NORMALISASI NAMA KOLOM

Dari dataframe yang ditampilkan di atas, kita memiliki satu kolom yang paling awal (paling kiri) yang tidak bernama (unnamed), kolom yang tidak bernama ini akan menyulitkan proses analisis data. Maka, ada dua alternatif solusi:

1. Menghapus (drop) kolom tersebut
2. Memberikan nama pada kolom tersebut

Dalam dataset, terlihat bahwa konteks untuk kolom pertama tersebut adalah ID dari lagunya. Maka, untuk sementara, kolom tersebut bisa diberi nama "ID"

```python
df.columns.tolist()
```

Untuk memastikan bahwa konteks kolom pertama tersebut benar adalah ID (unik untuk setiap lagu), maka kita bisa periksa apakah judul lagu yang sama memiliki ID yang sama. Sebagai contoh (dan agar penjelasan lebih intuitif) akan dianalisa ID dari lagu "Dance Monkey"

```python
df.loc[df["title"]=="Dance Monkey"]
```

Dari hasil di atas terlihat bahwa lagu yang sama, yaitu "Dance Monkey" ciptaan Tones and I, memiliki ID yang berbeda-beda, padahal secara intuitif ID lagu seharusnya unik untuk setiap lagu yang unik (berbeda), bukan untuk setiap entri yang berbeda. Oleh karena itu, dirasa kolom ID ini kurang cocok untuk konteks ID, lebih baik dinamakan kolom "nomor". Tetapi, mengingat bahwa dataframe sudah memiliki indeks, sebenarnya kolom "nomor" tidak diperlukan (redundant).

Dari pemaparan di atas, bisa disimpulkan bahwa kolom ID bisa dihapus karena tidak diperlukan.

## PEMERIKSAAN NILAI NULL

```python
df.loc[df.isnull().any(axis=1)]
```

Melihat dari hasil di atas, kolom dengan nilai kosong paling banyak adalah kolom "top genre", dengan 15 nilai kosong dan entri dengan atribut kosong paling banyak adalah entri pada index 749, dengan 8 nilai kosong.

Ada 2 opsi untuk mengatasi nilai kosong:
1. Menghapus entri/kolom yang memiliki nilai kosong
2. Mengisi nilai kosong tersebut dengan suatu nilai lain (imputing)

Untuk nilai kosong pada entri dengan index 749 lebih masuk akal dilakukan penghapusan, karena data yang dimiliki sangat banyak (1000 entri) dan menghapus satu entri untuk negara India tidak akan terlalu memengaruhi analisis.

Sedangkan untuk nilai kosong pada kolom "top genre" lebih masuk akal dilakukan pengisian karena dataset hanya memiliki 2 kolom atribut kategorikal, sehingga penghapusan kolom "top genre" akan mempersulit analisis karena kekurangan atribut kategorikal.

Penanganan akan dimulai dengan nilai kosong pada entri index 749, tentukan terlebih dahulu urutan entri (lagu) tersebut pada urutan popularitas di negara India.

```python
df_India = df.loc[df["country"]=="india"]
df_India.sort_values("pop", ascending=0)
```

dilihat dari output program di atas, entri dengan atribut NaN yang banyak adalah lagu dengan ranking terendah dari negara India. Jadi tidak ada urutan di bawah atau di atasnya yang akan terpengaruh bila dilakukan penghapusan. Melihat kasus ini, maka opsi penghapusan masuk akal untuk data ini. Penghapusan akan dilakukan di akhir, bersama dengan penghapusan data yang irelevan dengan analisis.

selanjutnya, untuk kolom "top genre", *imputing* akan menggunakan categorical impute pada dari modul sklearn_pandas, seperti berikut

```python
cat_imputer = CategoricalImputer()
df["top genre"] = pd.Series(cat_imputer.fit_transform(df["top genre"]))
df["top genre"].isnull().sum()
```

sekarang bisa kita cek apakah masih ada data NaN pada kolom "top genre"

```python
df.loc[df.isnull().any(axis=1)]
```

Bisa dilihat, telah berhasil dilakukan imputing untuk menggantikan nilai-nilai NaN pada kolom "top genre"

## MENGHAPUS ENTRI DAN ATRIBUT YANG IRELEVAN

```python
df
```

Data yang dimiliki sekarang terdiri dari 1000 entri dan 16 kolom, kita bisa analisa kolom-kolom dan entri-entri yang bisa dihapus berdasarkan:
1. Untuk entri, bisa dihapus bila dirasa data yang dimiliki terlalu banyak dan analisisnya sendiri tidak memerlukan data sebanyak itu
2. Untuk kolom kuantitatif, kolom dengan varians (variance) data yang rendah tidak menambahkan *insight* yang signifikan terhadap analisis.
3. Untuk kolom kategorikal, kolom dengan kategori entri paling seragam tidak memberikan *insight* yang signifikan terhadap analisis.

untuk entri, analis merasa bahwa top 50 lagu untuk setiap negara terlalu banyak, diputuskan untuk diambil setengahnya saja, sehingga akan difilter hanya top 25 lagu tiap negara.

selanjutnya untuk kolom, perhitungan variance akan digunakan method var() untuk pandas dataframe. Sedangkan keseragaman kategorinya akan dilihat dari value_counts()

### Kalkulasi varians

Pertama akan dihitung varians dari kolom kuantitatif, kolom "year" dikecualikan dalam kalkulasi karena analis akan memerlukan kolom tersebut sebagai salah satu kolom berisi *time series*

```python
variance_array = []
df_kuanti = df.iloc[:, 5:14]
for column in df_kuanti.columns.tolist():
  variance_array.append(df_kuanti[column].var())
variance_array
```

```python
np.mean(variance_array)
```

untuk melihat kolom mana yang perlu dihapus, akan digunakan rata-rata variansnya sebagai patokan. Kolom dengan varians di bawah rata-rata akan dihapus

```python
index_kolom_gagal = []
for variance in variance_array:
  if variance < np.mean(variance_array):
    index_kolom_gagal.append(variance_array.index(variance))

nama_kolom_gagal = []
for i in range(len(df_kuanti.columns.tolist())):
  if i in index_kolom_gagal:
    nama_kolom_gagal.append(df_kuanti.columns.tolist()[i])

nama_kolom_gagal
```

```python
df = df.drop(nama_kolom_gagal, axis=1)
df
```

Secara logika, cara paling mudah adalah melakukan penghapusan data-data dari negara Jepang dan Israel, tetapi sebelum hal tersebut bisa dilakukan, harus dipastikan bahwa ada cukup banyak data negara bila kedua negara tersebut dihapus.
